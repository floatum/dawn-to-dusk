```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import font_manager
from matplotlib.colors import LinearSegmentedColormap
import ipywidgets as widgets
from ipywidgets import HBox, VBox

from dashboard_constants import (
    BG, TEXT, TEXT_MUTED, GRID,
    PREDAWN, DAWN, MORNING, MIDDAY, AFTNOON, DUSK, NIGHT, DEEPNITE,
    SPECIES_PALETTE, COUNTY_SHORT, DEFAULT_SPECIES,
)
from dashboard_data import (
    load, hourly_from_totals, hourly_species_from_counts,
)

# design system
def _pick_font():
    available = {f.name for f in font_manager.fontManager.ttflist}
    for candidate in ("JetBrains Mono", "JetBrainsMono", "DejaVu Sans Mono"):
        if candidate in available:
            return candidate
    return "monospace"

FONT = _pick_font()

mpl.rcParams.update({
    "figure.facecolor":  BG,
    "axes.facecolor":    BG,
    "savefig.facecolor": BG,
    "font.family":       FONT,
    "font.size":         11,
    "text.color":        TEXT,
    "axes.labelcolor":   TEXT,
    "axes.titlecolor":   TEXT,
    "xtick.color":       TEXT_MUTED,
    "ytick.color":       TEXT_MUTED,
    "axes.edgecolor":    GRID,
    "grid.color":        GRID,
    "axes.spines.top":   False,
    "axes.spines.right": False,
})

def hour_gradient(n=24):
    stops = [
        (0, DEEPNITE), (4, PREDAWN), (6, DAWN),
        (9, MORNING), (12, MIDDAY), (15, AFTNOON),
        (18, DUSK), (20, NIGHT), (23, DEEPNITE),
    ]
    positions = [s[0] / 23 for s in stops]
    colors = [s[1] for s in stops]
    cmap = LinearSegmentedColormap.from_list("sun_moon", list(zip(positions, colors)), N=256)
    return [cmap(h / (n - 1)) for h in range(n)]

# load data ONCE at cell execution
DATA = load()

# ─── plot functions ──────────────────────────────────────────────────────────

def plot_clock(month_range, hour_range, county):
    fig, ax = plt.subplots(figsize=(6, 6), facecolor=BG,
                           subplot_kw={"projection": "polar"})
    ax.set_facecolor(BG)
    hourly = hourly_from_totals(DATA["totals"], month_range, hour_range, county)
    angles = np.array([2 * np.pi * h / 24 for h in range(24)])
    width = 2 * np.pi / 24 * 0.88
    colors = hour_gradient(24)
    heights = np.sqrt(hourly.values + 1)
    ax.bar(angles, heights, width=width, color=colors,
           edgecolor=BG, linewidth=0.8, align="center", zorder=3)
    ax.set_theta_zero_location("N")
    ax.set_theta_direction(-1)
    ax.set_xticks(angles)
    ax.set_xticklabels([f"{h:02d}" if h % 3 == 0 else "" for h in range(24)],
                       fontsize=9, color=TEXT)
    ymax = np.sqrt(max(hourly.max(), 1) * 1.1)
    ax.set_yticks([])
    ax.set_ylim(0, ymax)
    ax.grid(color=GRID, linewidth=0.4, alpha=0.6, zorder=1)
    ax.spines["polar"].set_color(GRID)
    ax.spines["polar"].set_linewidth(0.4)
    total = int(hourly.sum())
    ax.set_title(f"the birding clock  —  {total:,} observations",
                 color=TEXT, fontsize=13, fontweight="bold",
                 family=FONT, pad=16)
    plt.show()

def plot_county_circles(_unused):
    fig, ax = plt.subplots(figsize=(8, 6), facecolor=BG)
    ax.set_facecolor(BG)
    county_counts = DATA["county_counts"]
    bubbles = DATA["county_bubbles"]
    sorted_counties = county_counts.index.tolist()
    county_colors = hour_gradient(len(sorted_counties))
    for i, bubble in enumerate(bubbles):
        name = sorted_counties[i]
        count = county_counts[name]
        short = COUNTY_SHORT.get(name, name)
        ax.add_patch(plt.Circle((bubble.x, bubble.y), bubble.r,
                                color=county_colors[i], alpha=0.85, zorder=3))
        fs = max(7, min(13, int(bubble.r * 55)))
        ax.text(bubble.x, bubble.y + bubble.r * 0.12, short,
                ha="center", va="center", color=TEXT, fontsize=fs,
                fontweight="bold", family=FONT, zorder=4)
        ax.text(bubble.x, bubble.y - bubble.r * 0.18, f"{count:,}",
                ha="center", va="center", color=TEXT_MUTED,
                fontsize=max(6, fs - 2), family=FONT, zorder=4)
    ax.set_xlim(-1.1, 1.1)
    ax.set_ylim(-1.1, 1.1)
    ax.set_aspect("equal")
    ax.axis("off")
    ax.set_title("observations by county  —  where nl's birds are watched",
                 color=TEXT, fontsize=13, fontweight="bold", family=FONT, pad=12)
    plt.show()

def plot_species(month_range, hour_range, county, species_tuple):
    fig, ax = plt.subplots(figsize=(9, 4.5), facecolor=BG)
    ax.set_facecolor(BG)
    hours_day = list(range(max(5, hour_range[0]), min(20, hour_range[1]) + 1))
    if len(hours_day) < 2:
        ax.text(0.5, 0.5, "hour range too narrow",
                ha="center", va="center", transform=ax.transAxes,
                color=TEXT_MUTED, fontsize=11, family=FONT)
        ax.axis("off"); plt.show(); return
    if not species_tuple:
        ax.text(0.5, 0.5, "select one or more species",
                ha="center", va="center", transform=ax.transAxes,
                color=TEXT_MUTED, fontsize=11, family=FONT)
        ax.axis("off"); plt.show(); return
    total_per_hour = hourly_from_totals(
        DATA["totals"], month_range, hour_range, county
    ).reindex(hours_day, fill_value=0)
    any_drawn = False
    for i, sp_name in enumerate(species_tuple):
        color = SPECIES_PALETTE[i % len(SPECIES_PALETTE)]
        sp_counts = hourly_species_from_counts(
            DATA["species_counts"], sp_name, month_range, hour_range, county
        ).reindex(hours_day, fill_value=0)
        if sp_counts.sum() == 0 or total_per_hour.sum() == 0:
            continue
        rate = sp_counts / total_per_hour.replace(0, np.nan)
        if rate.max() == 0 or np.isnan(rate.max()):
            continue
        rel = (rate / rate.max()).fillna(0)
        ax.fill_between(hours_day, rel, alpha=0.22, color=color, zorder=2)
        ax.plot(hours_day, rel, color=color, linewidth=2.2,
                zorder=3, label=sp_name.lower())
        any_drawn = True
    if not any_drawn:
        ax.clear()
        ax.text(0.5, 0.5,
                "no observations of the selected species\nin this filter range",
                ha="center", va="center", transform=ax.transAxes,
                color=TEXT_MUTED, fontsize=11, family=FONT)
        ax.axis("off"); plt.show(); return
    ax.set_xlim(hours_day[0], hours_day[-1])
    ax.set_ylim(0, 1.15)
    ax.set_xticks(hours_day)
    ax.set_xticklabels([f"{h:02d}" for h in hours_day], color=TEXT_MUTED, fontsize=9)
    ax.set_yticks([0, 0.5, 1.0])
    ax.set_yticklabels(["0", "0.5", "1.0"], color=TEXT_MUTED, fontsize=8)
    ax.set_xlabel("hour of day", color=TEXT_MUTED, fontsize=10)
    ax.set_ylabel("relative activity", color=TEXT_MUTED, fontsize=10)
    ax.grid(axis="y", color=GRID, linewidth=0.4, alpha=0.6, zorder=1)
    ax.set_axisbelow(True)
    for spine in ax.spines.values():
        spine.set_color(GRID)
        spine.set_linewidth(0.4)
    ax.legend(loc="upper right", frameon=False, fontsize=9,
              labelcolor=TEXT, handlelength=1.5)
    ax.set_title("a day in the life  —  species activity curves",
                 color=TEXT, fontsize=13, fontweight="bold", family=FONT, pad=12)
    plt.show()

# controls
SLIDER_STYLE = {"description_width": "60px"}
WIDGET_WIDTH = "320px"

month_slider = widgets.IntRangeSlider(
    value=[1, 12], min=1, max=12, step=1,
    description="month", continuous_update=False,
    style=SLIDER_STYLE, layout=widgets.Layout(width=WIDGET_WIDTH),
)
hour_slider = widgets.IntRangeSlider(
    value=[0, 23], min=0, max=23, step=1,
    description="hour", continuous_update=False,
    style=SLIDER_STYLE, layout=widgets.Layout(width=WIDGET_WIDTH),
)
county_dropdown = widgets.Dropdown(
    options=[("all counties", "All")] +
            [(COUNTY_SHORT.get(c, c), c) for c in DATA["counties"]],
    value="All", description="county",
    style=SLIDER_STYLE, layout=widgets.Layout(width=WIDGET_WIDTH),
)
species_picker = widgets.SelectMultiple(
    options=DATA["species_pool"], value=DEFAULT_SPECIES,
    description="species", rows=10,
    style=SLIDER_STYLE, layout=widgets.Layout(width=WIDGET_WIDTH),
)

# wire
clock_out = widgets.interactive_output(
    plot_clock,
    {"month_range": month_slider, "hour_range": hour_slider, "county": county_dropdown},
)
county_out = widgets.interactive_output(
    plot_county_circles,
    {"_unused": hour_slider},
)
species_out = widgets.interactive_output(
    plot_species,
    {"month_range": month_slider, "hour_range": hour_slider,
     "county": county_dropdown, "species_tuple": species_picker},
)

# layout
title = widgets.HTML(
    value='<div style="font-family:monospace; color:#f4ecd8; padding:10px 0;">'
          '<div style="font-size:28px; font-weight:bold;">dawn to dusk — explorer</div>'
          '<div style="font-size:13px; color:#8a8578;">'
          'interactive dashboard · newfoundland birds · ebird 2025</div></div>'
)
controls_panel = VBox(
    [month_slider, hour_slider, county_dropdown, species_picker],
    layout=widgets.Layout(padding="10px"),
)
vizzes_panel = VBox([clock_out, county_out, species_out])
body = HBox([vizzes_panel, controls_panel])
app = VBox([title, body])
display(app)
```
